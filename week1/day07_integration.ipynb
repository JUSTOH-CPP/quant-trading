```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import sys
sys.path.append("../utils")

from indicators import compute_rsi, compute_macd, compute_bbands

print("All imports successful")
```

```python
def full_analysis(ticker, start="2019-01-01", end="2024-01-01"):
    """
    Complete Week 1 analysis pipeline for my ticker.
    Returns summary stats dict.
    """
    print(f"\n{'#'*50}")
    print(f" FULL ANALYSIS: {ticker}")
    print(f"{'#'*50}")

    # ---Data---
    df=yf.download(ticker, start=start, end=end, auto_adjust=True)
    df.columns=df.columns.get_level_values(0)
    price=df["Close"].squeeze()
    returns=np.log(price/price.shift(1)).dropna()

    # ---Core stats---
    ann_ret=returns.mean()*252
    ann_vol=returns.std()*np.sqrt(252)
    sharpe=(ann_ret-0.02)/ann_vol

    # ---Max drawdown---
    wealth=(1+returns).cumprod()
    drawdown=(wealth-wealth.cummax())/wealth.cummax()
    max_dd=drawdown.min()

    # ---Distribution---
    skew=returns.skew()
    kurt=returns.kurt()

    # --- VaR ---
    var_95 = returns.sort_values().iloc[int(0.05 * len(returns))]
    cvar_95 = returns.sort_values().iloc[int(0.05 * len(returns))].mean()

    # ---Current indicators ---
    rsi_val=compute_rsi(price).iloc[-1]
    macd_val, sig_val, _=compute_macd(price)
    macd_signal="BULLISH" if macd_val.iloc[-1]> sig_val.iloc[-1] else "BEARISH"

    # ---Print summary ---
    print(f"\n  PERFORMANCE")
    print(f"  Ann. Return    : {ann_ret:>8.2%}")
    print(f"  Ann. Volatility: {ann_vol:>8.2%}")
    print(f"  Sharpe Ratio   : {sharpe:>8.2f}")
    print(f"  Max Drawdown   : {max_dd:>8.2%}")
    print(f"\n  DISTRIBUTION")
    print(f"  Skewness       : {skew:>8.4f}")
    print(f"  Excess Kurtosis: {kurt:>8.4f}")
    print(f"  95% VaR        : {var_95:>8.2%}")
    print(f"  95% CVaR       : {cvar_95:>8.2%}")
    print(f"\n  CURRENT SIGNALS")
    print(f"  RSI(14)        : {rsi_val:>8.2f}")
    print(f"  MACD Signal    : {macd_signal:>8}")

    return {
        "ticker"  : ticker,
        "ann_ret" : ann_ret,
        "ann_vol" : ann_vol,
        "sharpe"  : sharpe,
        "max_dd"  : max_dd,
        "skew"    : skew,
        "kurt"    : kurt,
        "var_95"  : var_95,
        "rsi"     : rsi_val,
        "macd"    : macd_signal
    }
```

```python
spy_result=full_analysis("SPY")
```

```python
tickers = ["SPY", "AAPL", "BTC-USD", "GLD", "TLT"]
results= []

for t in tickers:
    r = full_analysis(t)
    results.append(r)
```

```python
df_results = pd.DataFrame(results)
df_results = df_results.set_index("ticker")
df_ranked = df_results.sort_values("sharpe", ascending=False)

print(f"\n{'='*65}")
print(f" 5-ASSET COMPARISON - RANKED BY SHARPE")
print(f"{'='*65}")
print(f"{'Asset':<12} {'Ann Ret':>9} {'Ann Vol':>9} {'Sharpe':>8} {'Max DD':>9} {'Kurt':>7}")
print("-"* 65)
for idx, row in df_ranked.iterrows():
    print(f"{idx:<12} {row['ann_ret']:>9.2%} {row['ann_vol']:>9.2%} {row['sharpe']:>8.2f} {row['max_dd']:>9.2%} {row['kurt']:>7.2f}")
print("=" *65)
```
